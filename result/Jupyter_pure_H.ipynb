```python
import numpy as np
import astropy.constants as constants
import matplotlib.pyplot as plt
```

```python




# 
file_path = '/home/yuanyuxuan/Magrathea/result/Structure_Jupyter_pure_H.txt'

# 
with open(file_path, 'r') as file:
    lines = file.readlines()

# 
index = []
radius = []
pressure = []
mass = []
density = []
temperature = []
phase = []

# 
for line in lines[1:]:
    # 
    parts = line.strip().split('\t')
    if len(parts) == 7:  # 7
        print(line)
        index.append(int(parts[0]))
        radius.append(float(parts[1]))
        pressure.append(float(parts[2]))
        mass.append(float(parts[3]))
        density.append(float(parts[4]))
        temperature.append(float(parts[5]))
        phase.append(parts[6])

# NumPy
index = np.array(index)
radius = np.array(radius)
pressure = np.array(pressure)
mass = np.array(mass)
density = np.array(density)
temperature = np.array(temperature)
phase = np.array(phase)

# 
print("Index:", index)
print("Radius (Earth):", radius)
print("P (GPa):", pressure)
print("M (Earth):", mass)
print("Density (g/cm^3):", density)
print("T (K):", temperature)
print("Phase:", phase)
```
